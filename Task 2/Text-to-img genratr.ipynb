```python
from diffusers import StableDiffusionPipeline
import torch

# Load the Stable Diffusion pipeline
pipeline = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5")
pipeline.to("cuda")  # Use GPU for faster inference
```

```python
def generate_image(prompt: str):
    """
    Generate an image based on the input text prompt using Stable Diffusion.
    """
    try:
        # Generate image from text
        image = pipeline(prompt).images[0]
        return image
    except Exception as e:
        print(f"Error generating image: {e}")
        return None
```

```python
import gradio as gr

# Define the Gradio interface
def gradio_interface(prompt):
    """
    Generate an image from a prompt and return it.
    """
    image = generate_image(prompt)
    return image

# Set up Gradio interface
interface = gr.Interface(
    fn=gradio_interface,
    inputs=gr.Textbox(lines=2, placeholder="Enter your text description here..."),
    outputs=gr.Image(),
    title="Text-to-Image Generator",
    description="Enter a description, and the model will generate an image for you."
)

# Launch the application
interface.launch()
```
